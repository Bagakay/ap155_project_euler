```python
import numpy as np
import matplotlib.pyplot as plt
def f_func(x) :
    return (1-np.cos(x))/(x**2)
def central(x):
    h = np.logspace (-1,-6,500)
    fdp = ((-4*np.sin(x))/(x**3))+(((-6*np.cos(x))+6)/(x**4))+((np.cos(x))/(x**2))
    fdpa = (f_func(x+h) - 2*f_func(x) + f_func(x-h))/(h**2)

    E = abs(fdpa-fdp)
    return E
        
    

    
        

 
```

```python
def errorplot(x):
    h = np.logspace(-1,-6,500)
    E = central(x)
    fig, ax = plt.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(h, E)
```

```python
errorplot(0.004)
plt.savefig("myplot.png")
```

```python

def f_func1(x) :
    return (2*(np.sin(x/2))**2)/(x**2)
def central1(x):
    h = np.logspace (-1,-6,500)
    fdp = ((-4*np.sin(x))/(x**3))+(((-6*np.cos(x))+6)/(x**4))+((np.cos(x))/(x**2))
    fdpa2 = (f_func1(x+h) - 2*f_func1(x) + f_func1(x-h))/(h**2)

    E1 = abs(fdpa2-fdp)
    return E1
        
    

    
        

 
```

```python
def errorplot1(x):
    h = np.logspace(-1,-6,500)
    E1 = central1(x)
    fig, ax = plt.subplots()
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.plot(h, E1)
errorplot1(0.004)
plt.savefig("myplot1.png")
```
