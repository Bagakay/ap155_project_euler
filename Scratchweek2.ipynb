```python
t = np.arange(0, 10, 0.05)
freq = 1.0

y = np.sin(2*np.pi*freq*t)

fig, ax = plt.subplots()
ax.plot(y)
print(fig)
```

```python
t = np.arange(0, 10, 0.05)
freq = 1.0

y = np.sin(2*np.pi*freq*t)

fig, ax = plt.subplots()

shifted_y = np.sin(2*np.pi*freq*t - 2.0)

ax.plot(t, y, label="not shifted")
ax.plot(t, shifted_y, label="is shifted")

ax.legend()
```

```python
t_1 = np.arange(0, 2, 0.0005)
freq_1 = 1.0
freq_2 = 4.0

y_1=np.sin(2*np.pi*freq_1*t_1)
y_2=np.sin(2*np.pi*freq_2*t_1)
y = y_1*y_2

fig, ax = plt.subplots()

ax.plot(t_1, y, label="not shifted")

ax.legend()
```
