```python
#this is number 1
my_num = 1000

start = 1
div3 = []
while start < my_num:
    remainder = start % 3
    if remainder == 0:
        div3.append(start) 
    start = start + 1

print(div3)
```

```python
start = 1
div5 = []
while start < my_num:
    remainder = start % 5
    if remainder == 0:
        div5.append(start)
    start = start + 1
    print(div5)
```
